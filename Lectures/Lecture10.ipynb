## Basics: 

    Signals are measured data correlated with time

    

### Generate a few signals to use throughout the example

amplitude vs. time

    x - discrete, real signal

    h - discrete, real signal

    y - discrete, real signal

    z - discrete, complex signal


```python
import numpy as np
np.random.seed(42)
t = np.arange(25)
x = [np.random.randint(-5, 5) for x in t]
h = [np.random.randint(0, 7) for x in range(3)]
y = [np.random.randint(-2, 8) for x in range(10)]
# q = 4 + 5j
# q = complex(4, 5)
# print(q.imag)
z = [complex(x, np.random.randint(-2,7)) for x in x]
print(z)
```

### Plotting Discrete Time Signals

```python
import matplotlib.pyplot as plt
plt.figure(1)
plt.plot(t, x)
plt.xlabel('time n')
plt.ylabel('x[n]')
plt.title('Plot of Amplitude vs. Time')
```

```python
# plt.figure(2)
fig, (ax1, ax2, ax3) = plt.subplots(3)
ax1.plot(t, [z.real for z in z])
ax1.set_ylabel('Real(z[n])')
ax1.set_xlabel('time n')
ax1.grid(True)

ax2.plot(t, [z.imag for z in z])
ax2.set_ylabel('Imaginary(z[n])')
ax2.set_xlabel('time n')
ax2.grid(True)

ax3.plot(t, [abs(z) for z in z])
ax3.set_ylabel('|z[n]|')
ax3.set_xlabel('time n')
ax3.grid(True)

```

## Convolution

https://dsp.stackexchange.com/questions/4723/what-is-the-physical-meaning-of-the-convolution-of-two-signals

https://en.wikipedia.org/wiki/Convolution

https://www.scaler.com/topics/numpy-convolve/

```python
# Convolve x and h

plt.clf()
plt.figure()
fig, (ax1, ax2, ax3) = plt.subplots(3)

ax1.plot(t, x)
ax1.set_xlabel('time n')
ax1.set_ylabel('x[n]')
ax1.set_title('Plot of Amplitude vs. Time')
ax1.grid(True)

# ax2.plot(t, h)
ax2.plot(np.arange(len(h)), h)
ax2.set_xlabel('time n')
ax2.set_ylabel('h[n]')
ax2.set_title('Plot of Amplitude vs. Time for h')
ax2.grid(True)

xh = np.convolve(x,h)
ax3.plot(np.arange(len(xh)), xh, color='orange')
ax3.set_xlabel('time n')
ax3.set_ylabel('x[n]*h[n]')
ax3.set_title('Plot of Convolved Signals')
ax3.grid(True)
```

## Fourier Transforms

https://pythonnumericalmethods.berkeley.edu/notebooks/chapter24.02-Discrete-Fourier-Transform.html

https://pythonnumericalmethods.berkeley.edu/notebooks/chapter24.04-FFT-in-Python.html

https://numpy.org/doc/stable/reference/routines.fft.html


- Discrete time Fourier Transform (DTFT)
- Discrete Fourier Transform (DFT)
- Fast Fourier Tansform (FFT)

```python
xFFT = np.fft.fft(x)
w = [x*(2*np.pi/len(xFFT)) for x in np.arange(len(xFFT))]

plt.figure()
fig, (ax1, ax2) = plt.subplots(2)
ax1.plot(w, abs(xFFT))
ax1.set_xlabel('$\omega$ [rad]')
ax1.set_ylabel('|X($\omega$)|')
ax1.grid(True)

ax2.plot(w, np.angle(xFFT))
ax2.set_xlabel('$\omega$ [rad]')
ax2.set_ylabel('Angle(X($\omega$)) [rad]')
ax2.grid(True)
```
