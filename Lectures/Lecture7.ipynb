# What about Greyscale? 

Greyscale images are flat matrices, which are x * y * v(value)

For Matplotlib:

    https://stackoverflow.com/questions/12201577/how-can-i-convert-an-rgb-image-into-grayscale-in-python

For OpenCV:

    

```python
# Plt
img_gray = Image.open('Tazzy.jpg').convert('L')
print(img_gray)
# img_gray = Image.open('Tazzy.jpg').convert('LA')
plt.imshow(img_gray)

# 'LA' indicates that there is an alpha or transparency layer included (img edicting from phones)
```

```python
#cv2
imgcv_gray = cv2.cvtColor(imgcv, cv2.COLOR_BGRA2GRAY)
print(imgcv_gray.shape)
plt.imshow(imgcv_gray)
print(np.max(imgcv_gray))
print(np.min(img_gray))
plt.imshow(imgcv_gray, cmap='gray')
```
